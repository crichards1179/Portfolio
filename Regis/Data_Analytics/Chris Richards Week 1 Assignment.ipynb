### Dataset Description
CMS State Summary of Inpatient Charge Data for Fiscal Year 2011  
This dataset is a csv file provided by the Center for Medicare & Medicaid Services.  
It relates to Medicare expenditures by state and Diagnositic Related Group (DRG).
DRGs are definined as a classification system which groups patients based on similar clinical diagnoses and procedures administered during hospital stays.  
  
  Columns include:  
  - DRG Definition - A numeric DRG group number plus description
  - Provider State - A state abbreviation where the patient was hospitalized
  - Total Discharges - A count of hospital discharges by all provider types
  - Average Covered Charges - The average dollar amount of all charges considered covered by CMS
  - Average Total Payments - The average dollar amount of payments including charges not covered by CMS
  - Average Medicare Payments - The average dollar amount of all payments made by CMS

#### Boxplot  
A sample of the original dataset was taken containing the first 500 records.  
The figure size was set using figsize= to 15 x 15.  
The boxplot function of the Seaborn package was used to create the boxplot with the following parameters:  
- X = Provider State  
- Y = DRG Definition 2 (which is an abbreviation of the full DRG column in order to remove the long description strings)  
  
Viewing the resulting plot, it's clear that most states fall into a similar range of values as the box portion of the plot indicating the Interquartile Range (the middle 50% of values) are alike.  Further this indicates the those states have the majority of their data points in the same DRGs.  
  
A handful of states, Florida for example, have plots which cover a larger range of DRGs.  This indicates the states medical usage (measured by DRG) is more variable then others.  They also have "whiskers" which are much longer then the other groups further indicating their higher variability.  
  
In contrast, one state, Idaho, has a much more compact plot indicating very little variability across DRGs.  
  
Median values, indicated by the bar in the box, are generally higher in the boxplot or centered, with the exception of Alaska.  Alaska's median value is skewed towards the low end of the box indicating the majority of their data points exist in the lower DRGs.  
  
Outliers appear clustered around the same general range of higher DRGs.


```python
# Boxplot
d_sample = d.head(500)
fig, ax = plt.subplots(figsize=(15,15))
sb.boxplot(x='Provider State',
           y="DRG Definition 2",
           data = d_sample)
```

####  Read in txt.  Import modules: operator, regular expressions, and string.  From the natural language tool kit, import stop words list.

```python
folder = 'C:\\Users\\a2glnzz\\Documents\\Regis\\Analytics\\week6\\'
f = open(folder + 'alice.txt', 'r').read()

import operator, re, string
from nltk.corpus import stopwords
```

#### Convert all text to lowercase.  Using regex, split the text into individual words and assign it to a variable.  Display the first 50 entries.

```python
f = f.lower()
alice = f
alice_cloud = str(alice)
alice = re.split(r'\W+', alice)
alice[:50]
```

#### Import the natural language toolkit.  Download the list of stopwords in English.  Select all words in the text that are not stopwords and assign it to the variable.  Display the first 50 words.

```python
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')
alice = [ j for j in alice if j not in stop]
alice[:50]
```

#### Select all words that are not related to Guttenberg Project language.  Note:  this is not strictly necessary as these words should have been removed during preprocessing.  Display the result.

```python
remove_list = ['gutenberg', 'ebook', 'www', 'online']
alice = [ j for j in alice if j not in remove_list]
alice[:50]
```

#### Sum each word.  Create an empty dictionary to hold the words and counts.  Iterate thru the list of words, counting each word and storing the result in the dictionary.  Sort the dictionary by number of occurrences (the value) and print the first 50 words with their counts.

```python
new_alice = {}
for new_word in alice:
    if new_word in new_alice:
        new_alice[new_word] += 1
    else:
        new_alice[new_word] = 1

sorted_new_alice = sorted(new_alice.items(), key=operator.itemgetter(1), reverse = True)
print(sorted_new_alice[:50])
```
